```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
```

```python
response=requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table = soup.find('table', {'class': 'wikitable sortable'})
```

```python
videos = []
```

```python
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    if len(cols) > 1:
        rank = cols[0].text.strip()
        name = cols[1].text.strip()
        name = ''.join(filter(lambda x: not x.isdigit(), name)).replace('[', '').replace(']', '').strip()
        artist = cols[2].text.strip()
        upload_date = cols[4].text.strip()
        views = cols[3].text.strip()
        videos.append({
            'Rank': rank,
            'Name': name,
            'Artist': artist,
            'Upload date': upload_date,
            'Views': views
        })
```

```python
for video in videos:
    print(video)
```

```python
response = requests.get(india_gdp_url)
```

```python
if response.status_code!= 200:
    print(F"Failed to fetch the URL: {india_gdp_url}")
    exit()
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
table = soup.find('table', {'id': 'table_id'})
```

```python
gdp = []
```

```python
for row in table.tbody.find_all('tr'):
    cols = row.find_all('td')
    rank = cols[0].text.strip()
    state = re.sub(r'\[\d+\]', '', cols[1].text.strip())
    gsdp_18_19 = cols[2].text.strip()
    gsdp_19_20 = cols[3].text.strip()
    share_18_19 = cols[4].text.strip()
    gdp_billion = cols[5].text.strip()
    gdp.append({
        'Rank': rank,
        'State': state,
        'GSDP(18-19)': gsdp_18_19,
        'GSDP(19-20)': gsdp_19_20,
        'Share(18-19)': share_18_19,
        'GDP($ billion)': gdp_billion
    })
```

```python
for gdps in gdp:
    print(gdps)
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://github.com/trending'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
repos = soup.find_all('article', class_='Box-row', limit=10)
```

```python
for repo in repos:
    title_elem = repo.find('h1')
    if title_elem is not None:
        title = title_elem.a['href'].strip('/')
    else:
        title = ''

    description = repo.find('p', class_='col-9').text.strip() if repo.find('p', class_='col-9') else ''
    contributors = repo.find('a', class_='Link--muted').text.strip() if repo.find('a', class_='Link--muted') else ''
    language = repo.find('span', itemprop='programmingLanguage')
    language = language.text.strip() if language else 'Not specified'
```

```python
    print(f'Repository title: {title}')
    print(f'Repository description: {description}')
    print(f'Contributor count: {contributors}')
    print(f'Language used: {language}\n')
```
